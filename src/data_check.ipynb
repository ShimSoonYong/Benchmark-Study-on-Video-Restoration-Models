# Fine-tune slices

```python
import os
import numpy as np
import matplotlib.pyplot as plt

z_window = 512
def plot_one_sample(path: str, H_range: tuple, z_slice: int) -> None:
    print(path)
    data = np.fromfile(path, dtype=np.float32).reshape(-1, 512, 512)
    print(data.shape)
    for i in range(*H_range):
        plt.figure(figsize=(15, 6))
        plt.imshow(data[z_slice - z_window:z_slice, i, :], cmap='gray')
        plt.axis('off')
        root = os.path.dirname(path) + "/" + path.split('/')[-1].split('.')[0]
        os.makedirs(root, exist_ok=True)
        plt.savefig(root+ "/" + f"{i}.png", bbox_inches='tight', pad_inches=0)
        plt.close()
```

# Check the validity

```python
z_slices = [2000,
                1870,
                1840,
                1880,
                1680,
                1880,
                1920,
                1720,
                1770,
                1880,
                1800,
                1760,
                1860,
                2140,
                2200,
                1880,
                1860,
                2060,
                2160,
                1940,]
```

```python
def plot_vertical_reverse(path:str, H_range:tuple, z_slice:float) -> None:
    data = np.fromfile(path, dtype=np.float32).reshape(-1, 512, 512)
    print(data.shape)
    for i in range(*H_range):
        plt.figure(figsize=(15, 6))
        plt.imshow(data[z_slice:z_slice - z_window:-1, i, :], cmap='gray')
        plt.axis('off')
        root = os.path.dirname(path) + "/" + path.split('/')[-1].split('.')[0]
        os.makedirs(root, exist_ok=True)
        plt.savefig(root+ "/" + f"{i}.png", bbox_inches='tight', pad_inches=0)
        plt.close()
```

```python
from glob import glob
globs = sorted(glob('/workspace/simulation_data/single_phase_raw/*.bin'))

for i, g in enumerate(globs):
    plot_vertical_reverse(path=g,
                          H_range=(250, 350),
                          z_slice=z_slices[i])
```
